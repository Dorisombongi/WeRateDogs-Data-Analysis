# Report: WeRateDogs act_report


## WeRateDogs Data

We have three datasets which were downloaded, cleaned then merged for analysis;
The twitter archive dataset which was downloaded and contains 2356 of the 5000+ tweets from twitter account named @dog_rates from 15 november 2015 to 1 august 2017, the images prediction dataset which was downloaded programmatically using requests and data converted from test file to a dataframe and the tweet json dataset which was downloaded by querying Twitter Api using tweepy and data extracted programmatically from the text file to a dataframe.

## WeRateDogs Data Wrangling

Before analysis the datasets were cleaned for better analysis. The datasets were assessed both visually and programmatically using pandas and in excel for quality and tidiness issues. Nine quality issues and three tidiness issues were found and resolved by cleaning them. After data cleaning and merging the three datasets there was 1941 tweets in the master dataset with good quality and tidy data.
